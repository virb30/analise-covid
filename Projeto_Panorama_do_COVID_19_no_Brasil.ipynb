COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Esse risco é ainda maior em se tratando da população idosa. Idosos, naturalmente, são mais sucetíveis a doenças infectocontagiosas, uma vez que o sistema imunológico enfraquece com a idade, favorecendo a manifestação de casos mais graves nessa população.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, iremos realizar uma análise sobre os dados públicos da doença.


## Pandemia

Após mais de 1 ano de decretada pandemia de COVID-19, em março de 2020, além das medidas recomendadas pela OMS - distanciamento social, higienização das mãos e uso de máscara - esperança se concentra na vacinação em massa da população para que a vida volte ao normal.

## Obtenção dos Dados

Os dados foram obtidos do repositório da *Our World In Data* (OWID) um projeto da [*Global Change Data Lab*](https://global-change-data-lab.org/), uma organização sem fins lucrativos baseada no Reino Unido. O objetivo desse projeto é fazer o conhecimento sobre grandes problemas acessível e compreensível. Como diz o slogan do projeto (em tradução livre): **Pesquisa e dados para fazer progresso frente aos maiores problemas do mundo**.

Segundo a documentação do *dataset*, disponível em: https://github.com/owid/covid-19-data/tree/master/public/data, esse conjunto contem dados históricos da pandemia de COVID-19 até a data de publicação, e conta com dados referentes a: vacinação, testes e confirmações, hospitalizações e UTI, casos confirmados, óbitos, taxa de reprodução e resposta do governo.

Analisar dados de uma nova doença é um grande desafio pois não se têm informações suficientes sobre a mesma, tais como: como é feito o diagnóstico, quais os sintomas, quais os testes existentes etc., com isso os dados relacionados à ela também tendem a sofrer diversas alterações ao longo do tempo, seja mudanças na metodologia, ou acréscimos de novas variáveis que eram irrelevantes no início e passaram a ser relevantes agora, entre outros.

Notamos essa mesma natureza evolutiva em nosso *dataset* que passou por diversas modificações, estruturais e metodológicas, desde seu primeiro registro. Todas as mudanças podem ser vistas [aqui](https://github.com/owid/covid-19-data/tree/master/public/data#changelog).


### Fonte dos dados

Apesar de disponibilizado pelo OWID, os dados são coletados a partir de múltiplas fontes:

* **Casos confirmados e mortes**: provenientes do repositório de dados da COVID-19 da Universidade John Hopkins (EUA), o conjunto de dados é atualizado diariamente. Importante ressalter que números negativos para casos e mortes são possíveis, quando algum país realiza correções nos dados históricos.
* **Hospitalizações e UTI**: dados provenientes do Centro Europeu de Controle e Prevenção de Doenças, para alguns países europeus e de fontes dos governos dos Estados Unidos, Canada, Argélia, Suíça, Israel e Reino Unido.
* **Testes para COVID-19**: esses dados são coletados pela equipe do OWID a partir de relatórios oficiais.
* **Vacinação contra COVID-19**: dados coletados a partir de relatórios oficiais.
* **Outras variáveis**: dados coletados de múltiplas fontes.

### Panorama do COVID-19 no mundo

```
# plotar gráfico de linhas com a evolução do total de mortes no mundo
fig, ax = plt.subplots(figsize=(10,8))

df.loc[df.location == "World", "total_cases"].plot(ax=ax, lw=3, color=TAGGED_COLOR)

ax.set_ylabel("Quantidade de casos (em milhões)", fontdict={'fontsize': 'large'})
ax.set_xlabel("Data", fontdict={'fontsize': 'large'})
ax.set_title("Evolução de casos por COVID-19 no Mundo", fontdict={'fontsize': 'x-large'}, loc='left', position=(-0.05, 1.1))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
```

#### Verificando quantidade de casos e mortes no mundo e por continente em 16/08/2021

Iniciaremos nossa análise extraindo o total de casos e mortes (em 16/08/2021) por localidade. Para isso iremos extrair todas as localidades cuja variável `iso_code` inicia com `OWID`. Pelo que vimos em nossa análise inicial, os dados agregados relacionados ao mundo e aos continentes estão em registros rotulados com esse prefixo.


```
[DEFAULT_COLOR]*5
```

```
# plotar gráfico de total de casos por continente
top5_cases_location = df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID'))].sort_values(by='total_cases', ascending=False)[1:6]

fig, ax = plt.subplots()

colors = [DEFAULT_COLOR]*5
colors[0] = TAGGED_COLOR

bars = ax.barh('location', 'total_cases', data=top5_cases_location, color=colors)
ax.invert_yaxis()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Total de casos por continente', loc='left', fontdict={'fontsize': 'x-large'}, position=(-0.01, 1.05))

plt.tight_layout()
```

```
top5_deaths_location = df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID'))].sort_values(by='total_deaths', ascending=False)[1:6]

fig, ax = plt.subplots()

colors = [DEFAULT_COLOR]*5
colors[0] = TAGGED_COLOR

ax.barh('location', 'total_deaths', data=top5_deaths_location, color=colors)
ax.invert_yaxis()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Total de mortes por continente', loc='left', fontdict={'fontsize': 'x-large'}, position=(-0.01, 1.05))

plt.tight_layout()
```

Vemos que o continente Asiático é o que possui maior número de casos, no entanto, o continente Europeu é o que possui maior número de mortes.

Como o COVID-19 é uma doença infectocontagiosa e que se manifesta de maneira mais grave na população idosa, nossa hipótese inicial é que o continente Europeu possui mais habitantes nessa condição, justificando a presença do continente nessa posição. Vamos verificar se nossa hipótese é válida analisando o percentual da população acima de 65 anos nos continentes.

Como não temos esse dado disponível para os registros dos continentes, iremos preenchê-lo com a média da variável `aged_65_older` dos países localizados nesses continentes.

```
# atualizar dados das variáveis aged_65_older dos continentes 
df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_EUR')), 'aged_65_older'] = df.loc[(df.index == '2021-08-16') & (df.continent == 'Europe')].aged_65_older.mean()
df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_ASI')), 'aged_65_older'] = df.loc[(df.index == '2021-08-16') & (df.continent == 'Asia')].aged_65_older.mean()
df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_EUN')), 'aged_65_older'] = df.loc[(df.index == '2021-08-16') & (df.continent == 'European Union')].aged_65_older.mean()
df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_NAM')), 'aged_65_older'] = df.loc[(df.index == '2021-08-16') & (df.continent == 'North America')].aged_65_older.mean()
df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_SAM')), 'aged_65_older'] = df.loc[(df.index == '2021-08-16') & (df.continent == 'South America')].aged_65_older.mean()
```

```
# plotar gráfico dos continentes com maior parcela da população idosa (acima de 65 anos)
top5_aged_65_older_location = df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID'))].sort_values(by='aged_65_older', ascending=False)[:5]

fig, ax = plt.subplots()
colors = [DEFAULT_COLOR]*5
colors[0] = TAGGED_COLOR

ax.bar('location', 'aged_65_older', data=top5_aged_65_older_location, color=colors)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Média da população Idosa por continente', loc='left', fontdict={'fontsize': 'x-large'}, position=(-0.01, 1.05))

plt.tight_layout()
```

Apesar de impreciso (pois utilizamos a média da população idosa dos países) podemos ver que há uma disparidade entre o percentual da população Idosa no continente Europeu em relação aos demais continentes, essa característica é um forte indício que confirma nossa hipótese que justifica o maior número de mortes nesse continente.

#### Verificar os países com maior número de casos e mortes em 16/08/2021

#### O impacto da vacinação


> As vacinas para COVID-19 são efetivas e uma ferramenta importante para controle da pandemia. No entanto, nenhuma vacina é 100% eficaz para prevenir pessoas de contrairem a doença, existe uma pequena porcentagem de pessoas totalmente vacinadas que ainda ficam doentes (OPAS, 2021).

Portanto o objetivo central da vacinação em massa é a redução no número de casos graves e óbitos e, consequentemente, o controle da pandemia.  

Abaixo vemos como o avanço da vacinação impactou no número de novos casos e novas mortes no mundo. Consideraremos o número absoluto de pessoas vacinadas (com 1 ou todas as doses previstas).

Vale lembrar que cada país tem seu ritmo de vacinação e disponibilidade de vacinas, ou seja, o cenário mundial pode não se repetir quando analisamos os países isoladamente.

```
# plotar gráfico de pessoas vacinadas / novas mortes
fig, ax = plt.subplots(figsize=(15,8))

ax_2 = ax.twinx()
df.loc[df.location == "World", "new_deaths"].plot(ax=ax, lw=2, color=DEFAULT_COLOR)
df.loc[df.location == "World", "people_vaccinated"].plot(ax=ax_2, lw=3, color=TAGGED_COLOR)

ax.set_xlabel('Data')

ax.set_title("Pessoas vacinadas / Novas mortes no mundo", fontdict={'fontsize': 'x-large'}, loc='left', position=(-0.05, 1.05))
ax.set_ylabel("Novas mortes", fontdict={'fontsize': 'large'})

ax_2.tick_params(axis='y', labelcolor=TAGGED_COLOR)
ax_2.set_ylabel("Pessoas vacinadas", color=TAGGED_COLOR, fontdict={'fontsize': 'large'})

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax_2.spines['top'].set_visible(False)
ax_2.spines['right'].set_visible(False)

plt.tight_layout()
```

Vemos acima que o avanço da vacinação pode ter contribuído para a redução do número de novas mortes ao longo tempo.

No entanto, quando comparamos com o número de novos casos, não temos o mesmo resultado:

```
# plotar gráfico de pessoas vacinadas / novos casos
fig, ax = plt.subplots(figsize=(15,8))

ax_2 = ax.twinx()
df.loc[df.location == "World", "new_cases"].plot(ax=ax, lw=2, color=DEFAULT_COLOR)
df.loc[df.location == "World", "people_vaccinated"].plot(ax=ax_2, lw=3, color=TAGGED_COLOR)

ax.set_xlabel('Data')

ax.set_title("Pessoas vacinadas / Novos casos no mundo", fontdict={'fontsize': 'x-large'}, loc='left', position=(-0.05, 1.05))
ax.set_ylabel("Novos casos", fontdict={'fontsize': 'large'})

ax_2.tick_params(axis='y', labelcolor=TAGGED_COLOR)
ax_2.set_ylabel("Pessoas vacinadas", color=TAGGED_COLOR, fontdict={'fontsize': 'large'})

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax_2.spines['top'].set_visible(False)
ax_2.spines['right'].set_visible(False)

plt.tight_layout()
```

Notamos pelo gráfico acima uma oscilação na quantidade de novos casos com picos e valies cada 2 meses aproximadamente. Esse comportamente se repete, independente do avanço da vacinação. No entanto, a partir de Janeiro de 2021 a tendência de novos casos está praticamente constante.

Diversos fatores podem contribuir para esse cenário, por exemplo, adoção de medidas mais restritivas, ou ainda aumento da confiança da população já vacinada podem ter consequências no aparecimento de novos casos.

#### Verificando quantidade de casos e mortes nos países da América do Sul em 16/08/2021

Conforme vimos anteriormente, o Brasil é um dos países que mais registrou casos e mortes até 16/08/2021 no mundo, consequentemente, é o país da América do Sul que mais registrou casos e mortes no período.

Vamos verificar qual a participação do Brasil para números de casos e mortes para o continente Sul Americano.

```
sam_total_cases = df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_SAM'))].total_cases.values[0]
brazil_total_cases = df_brazil.loc[df_brazil.index == '2021-08-16'].total_cases.values[0]

print('Relação Brasil / América do Sul em número de casos:')
print(f'Total de casos na América do Sul em 16/08/2021: {sam_total_cases}')
print(f'Total de casos no Brasil em 16/08/2021: {brazil_total_cases}')
print(f'Brasil/América do Sul: {brazil_total_cases / sam_total_cases * 100:.2f}%')

sam_total_deaths = df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_SAM'))].total_deaths.values[0]
brazil_total_deaths = df_brazil.loc[df_brazil.index == '2021-08-16'].total_deaths.values[0]

print('\nRelação Brasil / América do Sul em número de mortes:')
print(f'Total de mortes na América do Sul em 16/08/2021: {sam_total_deaths}')
print(f'Total de mortes no Brasil em 16/08/2021: {brazil_total_deaths}')
print(f'Brasil/América do Sul: {brazil_total_deaths / sam_total_deaths * 100:.2f}%')

sam_population = df.loc[(df.index == '2021-08-16') & (df.iso_code.str.startswith('OWID_SAM'))].population.values[0]
brazil_population = df_brazil.loc[df_brazil.index == '2021-08-16'].population.values[0]

print('\nRelação Brasil / América do Sul em população total:')
print(f'População Total América do Sul em 16/08/2021: {sam_population}')
print(f'População Total Brasil em 16/08/2021: {brazil_population}')
print(f'Brasil/América do Sul: {brazil_population / sam_population * 100:.2f}%')
```
